```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_ce = pd.read_csv("./log/baseline_ce.csv")
df_mae = pd.read_csv("./log/baseline_mae.csv")
df_focal_5 = pd.read_csv("./log/baseline_focal0.5.csv")
df_focal_2 = pd.read_csv("./log/baseline_focal2.csv")

df_ce_r = pd.read_csv("./log/resnet18_ce.csv")
df_mae_r = pd.read_csv("./log/resnet18_mae.csv")
df_focal_5_r = pd.read_csv("./log/resnet18_focal0.5.csv")
df_focal_2_r = pd.read_csv("./log/resnet18_focal2.csv")
```

```python
# baseline
fig, ax = plt.subplots()
ax.plot(df_mae["f1score_macro"], label="mae")
ax.plot(df_ce["f1score_macro"], label="ce")
ax.plot(df_focal_5["f1score_macro"], label="focal0.5")
ax.plot(df_focal_2["f1score_macro"], label="focal2")
ax.set_xlabel("Epoch number")
ax.set_ylabel("Macro F1-Score")
plt.title("Base Model")
plt.legend()
plt.savefig("./figures/baseline_model.pdf")
```

```python
# baseline_acc
fig, ax = plt.subplots()
ax.plot(df_mae["acc"], label="mae")
ax.plot(df_ce["acc"], label="ce")
ax.plot(df_focal_5["acc"], label="focal0.5")
ax.plot(df_focal_2["acc"], label="focal2")
ax.set_xlabel("Epoch number")
ax.set_ylabel("Accuracy")
plt.title("Base Model")
plt.legend()
plt.savefig("./figures/baseline_model_acc.pdf")
```

```python
# resnet18_acc
fig, ax = plt.subplots()
ax.plot(df_mae_r["f1score_macro"], label="mae")
ax.plot(df_ce_r["f1score_macro"], label="ce")
ax.plot(df_focal_5_r["f1score_macro"], label="focal0.5")
ax.plot(df_focal_2_r["f1score_macro"], label="focal2")
ax.set_xlabel("Epoch number")
ax.set_ylabel("Macro F1-Score")
plt.title("Resnet18")
plt.legend()
plt.savefig("./figures/resnet18_model.pdf")
```

```python
# resnet18_acc
fig, ax = plt.subplots()
ax.plot(df_mae_r["acc"], label="mae")
ax.plot(df_ce_r["acc"], label="ce")
ax.plot(df_focal_5_r["acc"], label="focal0.5")
ax.plot(df_focal_2_r["acc"], label="focal2")
ax.set_xlabel("Epoch number")
ax.set_ylabel("Accuracy")
plt.title("Resnet18")
plt.legend()
plt.savefig("./figures/resnet18_model_acc.pdf")
```
