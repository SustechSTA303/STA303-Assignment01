```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.arange(0, 1, 0.0001)
y1 = (1 - x) ** 0.5
y2 = (1 - x) ** 2
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, color="red")
ax.plot(x, y2, color="darkcyan")
plt.legend([r"$(1-p)^\gamma, \gamma=0.5$ ", r"$(1-p)^\gamma, \gamma=2$"])
plt.xlabel(r"$p$")
plt.ylabel(r"$(1-p)^\gamma$")
plt.savefig("./figures/focal1.pdf")
```

```python
epsilons = [1e-1, 1e-2, 1e-9]
fig, ax = plt.subplots()
x2 = torch.arange(0, 1.1, 0.1)
for epsilon in epsilons:
    y = (1-x2+epsilon) ** 0.5
    ax.scatter(x2, y)
ax.plot(x, y1, color="red")
plt.xlabel(r"$p$")
plt.ylabel(r"$(1-p+\epsilon)^{0.5}$")
plt.legend([r"$\epsilon=0.1$", r"$\epsilon=0.01$", r"$\epsilon=1\times 10^{-9}$", r"$\epsilon=0$"])
plt.savefig("./figures/focal2.pdf")
```
