```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
x = symbols('x')
x_values = np.arange(start=0.01, stop=0.99, step=0.01)
```

```python
# 定义函数
func_mae = 1-x
func_ce = -log(x)
func_focal5 = -(1-x)**0.5 * log(x)
func_focal2 = -(1-x)**2 * log(x)

# 求导数
derivative_mae = diff(func_mae, x)
derivative_ce = diff(func_ce, x)
derivative_focal5 = diff(func_focal5, x)
derivative_focal2 = diff(func_focal2, x)

# 计算函数和导数值
mae_values = [func_mae.subs(x, i) for i in x_values]
ce_values = [func_ce.subs(x, i) for i in x_values]
focal5_values = [func_focal5.subs(x, i) for i in x_values]
focal2_values = [func_focal2.subs(x, i) for i in x_values]

derivative_values_mae = [derivative_mae.subs(x, i) for i in x_values]
derivative_values_ce = [derivative_ce.subs(x, i) for i in x_values]
derivative_values_focal5 = [derivative_focal5.subs(x, i) for i in x_values]
derivative_values_focal2 = [derivative_focal2.subs(x, i) for i in x_values]

fig, ax = plt.subplots()
ax.plot(x_values, derivative_values_mae, label=r"$\mathcal{D}\ MAE(p, 1)$")
ax.plot(x_values, derivative_values_ce, label=r"$\mathcal{D}\ CE(p, 1)$")
ax.plot(x_values, derivative_values_focal5, label=r"$\mathcal{D}\ Focal(p, 1, \gamma=0.5)$")
ax.plot(x_values, derivative_values_focal2, label=r"$\mathcal{D}\ Focal(p, 1, \gamma=2)$")

plt.legend()
plt.title(r"$\mathcal{D\ loss}$")
plt.savefig("./figures/losses_derivative.pdf")
```

```python
fig, ax = plt.subplots()
ax.plot(x_values, mae_values, label=r"$MAE(p, 1)$")
ax.plot(x_values, ce_values, label=r"$CE(p, 1)$")
ax.plot(x_values, focal5_values, label=r"$Focal(p, 1, \gamma=0.5)$")
ax.plot(x_values, focal2_values, label=r"$Focal(p, 1, \gamma=2)$")

plt.legend()
plt.title(r"$\mathcal{loss}$")
plt.savefig("./figures/losses_values.pdf")
```
